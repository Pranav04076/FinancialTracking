```python
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.linear_model import LogisticRegression

from sklearn.pipeline import Pipeline

from sklearn.metrics import accuracy_score

import joblib
```

```python
df1 = pd.read_csv("../../TestDataSet/categorization/combined_expense_dataset.csv")
```

```python
df2 = pd.read_csv("../../TestDataSet/categorization/personal_expense_classification.csv")
```

```python
df3 = pd.read_excel("../../TestDataSet/categorization/expense_dataset_30000_realistic.xlsx")

df = pd.concat([df1, df2, df3], ignore_index = True)

df = df.drop_duplicates()
```

```python
df['category'].value_counts()
```

```python
df["category"] = df["category"].str.strip()
```

```python
category_mapping = {
    "food": "Food & Dining",
    "transport": "Transport",
    "shopping": "Shopping",
    "entertainment": "Entertainment",
    "technology": "Electronics"
}

df["category"] = df["category"].replace(category_mapping)
```

```python
x = (
    df["merchant"].fillna("")
    + " "
    + df["description"].fillna("")
)
y=df['category']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)
```

```python
model = Pipeline([
        ("tfidf", 
         TfidfVectorizer()
        ),
        ("classifier", 
         LogisticRegression(max_iter=1000)
        )
])
```

```python
model.fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, predictions)
print(accuracy)
```

```python
test_merchants = [
    "Dominos",
    "Pizza Hut",
    "McDonalds",
    "KFC",
    "Starbucks",
    "IRCTC",
    "BookMyShow"
]

for m in test_merchants:
    print(m, "->", model.predict([m])[0])
```

```python
import joblib
joblib.dump(model, "../ml/model.pkl")
```
